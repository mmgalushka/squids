# Synthetic images of specific shapes

This demonstrates generating synthetic images of specific shapes:
- 'r' - rectangles;
- 't' - triangles;
- 'e' - ellipses.

For example to generate images containing all three shapes `adding_shapes` mas be set to `rte`, for generating only rectangular shapes to `r`, etc.

```python
# Let's make sure that JNB can find our SquiDS package ;)
import sys
sys.path.insert(0, '..')

# The random package is needed to set a random state for
# generating consistent images. 
import random

from PIL import ImageDraw

from squids.dataset.image import create_synthetic_image
from squids.dataset.palette import Palette
from squids.dataset.background import Background
```

## Generate image with combination of all shapes

```python
random.seed(365)
image, shapes = create_synthetic_image(
    image_width=256,
    image_height=256,
    image_palette=Palette.RGB,
    image_background=Background.BLACK,
    image_capacity=3,
    adding_shapes="rte",
    noise_level=0)
print("Generated image shapes:", list(map(str, shapes)))
draw = ImageDraw.Draw(image)
draw.rectangle([0,0, 255, 255])
image.show()
```
